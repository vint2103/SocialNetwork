```python
import networkx as nx
from networkx.algorithms.community import girvan_newman
import matplotlib.pyplot as plt
```

```python
G=nx.les_miserables_graph()
print(f"sonut: {G.number_of_nodes()}")
print(f"socanh: {G.number_of_edges()}")
communities=girvan_newman(G)

communities=list(communities)

print(f"solan lap toi da: {len(communities)}")
```

```python
def draw_graph(G, show_names = False, node_size = 50):
    pos = nx.spring_layout(G)
    nx.draw(G, pos, with_labels = show_names, node_size = node_size, font_size = 8)
    plt.show()
```

```python
selected_comunities = communities[9]
```

```python
for i, communities in enumerate(selected_comunities):
    g_communities = G.subgraph(communities)
    print(f"Community: {i}:{communities}")
    draw_graph(g_communities, show_names=True, node_size=50)
```
