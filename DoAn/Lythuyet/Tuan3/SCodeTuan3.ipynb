```python
G_lesmis = nx.les_miserables_graph()
print(f"S nt: {G_lesmis.number_of_nodes()}")
print(f"S cnh: {G_lesmis.number_of_edges()}")
```

```python
pos_lesmis = nx.circular_layout(G_lesmis)
plt.figure(figsize=(10, 10))
nx.draw(G_lesmis, pos_lesmis, with_labels=True, node_size=500, edge_color="darkgreen")
plt.title("Karate KID")
plt.show()
```

```python
density = nx.density(G_lesmis)
density
```

```python
deg_cent = nx.degree_centrality(G_lesmis)
degcent_df = pd.DataFrame(deg_cent, index=[0]).T
degcent_df.columns = ["degree_centrality"]
degcent_df = degcent_df.sort_values(by="degree_centrality", ascending=False)
title = "Top 20:"
_ = degcent_df[0:20].plot.barh(title=title, figsize=(12, 7))
plt.gca().invert_yaxis()
```

```python
betwcent = nx.betweenness_centrality(G_lesmis)
betwcent_df = pd.DataFrame(betwcent, index=[0]).T
betwcent_df.columns = ["betweenness_centrality"]
betwcent_df.sort_values("betweenness_centrality", inplace=True, ascending=False)
title = "Top 20"

# To biu  ct ngang (barh) vi 20 ti khon u tin trong DataFrame `betwcent_df`, vi tiu  v kch thc (12, 7)
_ = betwcent_df[0:20].plot.barh(title=title, figsize=(12, 7))

# o ngc trc y  cc gi tr cao hn nm trn cng
plt.gca().invert_yaxis()
```

```python
closecent = nx.closeness_centrality(G)
# Chuyn i t dictionary `closecent` sang DataFrame v chuyn i index thnh cc keys ca dictionary
closecent_df = pd.DataFrame(closecent, index=[0]).T

# t tn ct duy nht l 'closeness_centrality'  biu din gi tr  trung tm gn
closecent_df.columns = ["closeness_centrality"]

# Sp xp cc hng trong DataFrame theo gi tr ct 'closeness_centrality' theo th t gim dn
closecent_df.sort_values("closeness_centrality", inplace=True, ascending=False)
title = "Top 20"

# To biu  ct ngang (barh) vi 20 ti khon u tin trong DataFrame `betwcent_df`, vi tiu  v kch thc (12, 7)
_ = closecent_df[0:20].plot.barh(title=title, figsize=(12, 7))

# o ngc trc y  cc gi tr cao hn nm trn cng
plt.gca().invert_yaxis()
```

```python
pagerank = nx.pagerank(G_lesmis)

pagerank_df = pd.DataFrame(pagerank, index=[0]).T
pagerank_df.columns = ["pagerank"]
pagerank_df.sort_values("pagerank", inplace=True, ascending=False)

# t tiu  biu  l 'Top 20'
title = "Top 20"

# To biu  ct ngang (barh) vi 20 ti khon u tin trong DataFrame `betwcent_df`, vi tiu  v kch thc (12, 7)
_ = pagerank_df[0:20].plot.barh(title=title, figsize=(12, 7))

# o ngc trc y  cc gi tr cao hn nm trn cng
plt.gca().invert_yaxis()
```

```python
density = nx.density(G_lesmis)
degrees = dict(G_lesmis.degree())
avg_degree = sum(degrees.values()) / len(G_lesmis)
adj_matrix = nx.adjacency_matrix(G_lesmis).todense()

# Visualize degree distribution
plt.figure(figsize=(8, 6))
plt.hist(list(degrees.values()), bins=20)
plt.title("Degree Distribution")
plt.xlabel("Degree")
plt.ylabel("Count")
plt.show()

# Calculate centrality measures
deg_cent = nx.degree_centrality(G_lesmis)
close_cent = nx.closeness_centrality(G_lesmis)
bet_cent = nx.betweenness_centrality(G_lesmis)

# Create a summary DataFrame
summary = pd.DataFrame(
    {
        "node": list(G_lesmis.nodes()),
        "degree": [degrees[node] for node in G_lesmis.nodes()],
        "deg_cent": [deg_cent[node] for node in G_lesmis.nodes()],
        "close_cent": [close_cent[node] for node in G_lesmis.nodes()],
        "bet_cent": [bet_cent[node] for node in G_lesmis.nodes()],
    }
)
summary = summary.sort_values("bet_cent", ascending=False)
print(summary.head())

# Visualize the network
pos = nx.spring_layout(G_lesmis)
plt.figure(figsize=(12, 12))
nx.draw(
    G_lesmis,
    pos_lesmis,
    with_labels=True,
    node_size=[v * 1000 for v in deg_cent.values()],
    cmap=plt.cm.Blues,
)
plt.title("Karate Club Network")
plt.show()
```
